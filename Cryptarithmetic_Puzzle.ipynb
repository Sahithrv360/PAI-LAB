4. Cryptarithmetic Puzzle
Solve SEND + MORE = MONEY where each letter represents a unique digit 0-9. 
Variables are letters (S,E,N,D,M,O,R,Y);domains are digits 0-9; 
constraints ensure unique digits and valid arithmetic (carry-over rules).

```python
class IncompatibleException(Exception):
    pass
```

```python
def SolveCryptArithmatic(word1,word2,result_word):
        letters = set(word1 + word2 + result_word)

        if len(letters) > 10 :
            exit()

        domains = {ch:set(range(10)) for ch in letters}

        for ch in {word1[0],word2[0],result_word[0]} :
            domains[ch].discard(0)

        def word_to_number(word,assignment):
            return int("".join(str(assignment[ch]) for ch in word))
        
        def valid_sum(assignment):

            try :
                return (
                        word_to_number(word1,assignment)+
                        word_to_number(word2,assignment) == 
                        word_to_number(result_word,assignment)
                        )
            except KeyError:
                return False

        
        def dfs(assignment):
            
            if len(assignment) == len(domains):
                return assignment if valid_sum(assignment) else None
            
            var = next(ch for ch in domains if ch not in assignment)

            for value in domains[var]:
                if value not in assignment.values():
                    assignment[var] = value

                    result = dfs(assignment)

                    if result:
                        return result
                    del assignment[var]

        solution = dfs({}) 
        if solution:
            print('\n solution found : ')
            for k in sorted(solution):
                print(f'{k} = {solution[k]}')
        else :
            return None
        print(f'\n{word_to_number(word1,solution)} + '
                f'{word_to_number(word2,solution)} ='
                f'{word_to_number(result_word,solution)}')
    
```

```python
def CryptArithematicInputs():
    word1 = input('Enter Word 1 : ').upper()
    word2 = input('Enter Word 2 : ').upper()
    result_word = input('Enter Result Word : ').upper()

    SolveCryptArithmatic(word1,word2,result_word)


CryptArithematicInputs()
```
