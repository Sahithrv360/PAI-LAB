```python
def label_the_jugs_in_litres():
    cap1 = int(input('Enter the capacity of Jug - 1 in lits : '))
    cap2 = int(input('Enter the capacity of Jug - 2 in lits : '))
    target = int(input('Enter the target capacity of Jug - 1 in lits : '))

    Solution(cap1, cap2, target)
```

```python
def Solution(cap1 ,cap2 ,target):

    visited = set()
    path = []
    
    def cal_time():
        pass
    def dfs(jug1,jug2):
        if (jug1,jug2) in visited:
            return False

        visited.add((jug1,jug2))
        path.append((jug1,jug2))

        if jug1 == target or jug2 == target:
            return True

        if dfs(cap1,jug2) :
            return True

        if dfs(jug1,cap2):
            return True

        if dfs(0,jug2):
            return True

        if dfs(jug1,0):
            return True

        # pour jug2 -> jug1
        transfer = min(jug2,cap1-jug1)
        if dfs(jug1+transfer,jug2-transfer) :
            return True
            
        # pour jug1 -> jug2    
        transfer = min(jug1,cap2-jug2)
        if dfs(jug1-transfer,jug2+transfer) :
            return True

        path.pop()
        return False
        
    dfs(0,0)
    if not path:
        print('No Solution is possible .')
    else :
        print('The path followed : ')
        for _ in path:
            print(f'{_}')
            
        visual_representation(path)
        simple_repre(path)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def visual_representation(path):
    G = nx.DiGraph()

    for i in range(len(path) - 1):
        G.add_edge(path[i],path[i+1])

    pos = nx.spring_layout(G)
    plt.figure(figsize = (6,6))

    nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1500, font_size=12, font_weight='bold')
    nx.draw_networkx_edges(G, pos, edgelist=list(G.edges()), edge_color='black', width=2)

    plt.title("Water Jug Problem - DFS Solution Path")
    plt.show()
```
