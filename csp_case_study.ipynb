```python
class ProctorAssignmentCSP:
    def __init__(self, exams, proctors):
        self.exams = exams
        self.proctors = proctors
        self.assignment = {}

    def is_valid(self, exam, proctor):
        # Skill constraint
        if exam["subject"] not in proctor["skills"]:
            return False
        
        # Availability constraint
        if exam["time"] not in proctor["available_times"]:
            return False
        
        # Time conflict constraint
        for assigned_exam_name, assigned_proctor_name in self.assignment.items():
            
            if assigned_proctor_name == proctor["name"]:
                
                # Find the exam object using its name
                for e in self.exams:
                    if e["name"] == assigned_exam_name:
                        if e["time"] == exam["time"]:
                            return False
        
        return True

    def backtrack(self, index=0):
        if index == len(self.exams):
            return True
        
        exam = self.exams[index]

        for proctor in self.proctors:
            if self.is_valid(exam, proctor):
                self.assignment[exam["name"]] = proctor["name"]
                
                if self.backtrack(index + 1):
                    return True
                
                # Backtrack step
                del self.assignment[exam["name"]]
        
        return False

    def solve(self):
        if self.backtrack():
            return self.assignment
        else:
            return None
```

```python
exams = [
    {"name": "Math Exam", "time": "9AM", "subject": "Math"},
    {"name": "CS Exam", "time": "9AM", "subject": "CS"},
    {"name": "Physics Exam", "time": "11AM", "subject": "Physics"}
]

proctors = [
    {"name": "Alice", "available_times": ["9AM", "11AM"], "skills": ["Math", "Physics"]},
    {"name": "Bob", "available_times": ["9AM"], "skills": ["CS"]},
    {"name": "Charlie", "available_times": ["11AM"], "skills": ["Physics", "CS"]}
]

csp = ProctorAssignmentCSP(exams, proctors)
solution = csp.solve()

if solution:
    print("Valid Assignment Found:\n")
    for exam, proctor in solution.items():
        print(f"{exam} → {proctor}")
else:
    print("No valid assignment possible.")
```
