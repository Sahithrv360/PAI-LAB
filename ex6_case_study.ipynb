```python
from pgmpy.models import DiscreteBayesianNetwork as DBN
from pgmpy.factors.discrete import TabularCPD as TCPD
from pgmpy.inference import VariableElimination as VE
```

```python
model = DBN([
             ('EarthQuake','Tornado'),
             ('EarthQuake','AlarmRinging'),
             ('Tornado','AlarmRinging'),
             ('Burglary','AlarmRinging'),
    ])

EQ = TCPD(variable='EarthQuake',variable_card=2,values=[[0.4],[0.6]])

Tor = TCPD(variable='Tornado',variable_card=2,evidence=['EarthQuake'], evidence_card=[2],
           values=[[0.8,0.5],[0.2,0.5]])

Bur = TCPD(variable='Burglary',variable_card=2,values=[[0.7],[0.3]])

AR = TCPD(variable='AlarmRinging',variable_card=2,evidence=['EarthQuake','Burglary','Tornado'],
           evidence_card=[2,2,2],
          values=[[1,.9,.95,.85,.89,.7,.87,.3],[0,.1,.05,.15,.11,.3,.13,.7]])
```

```python
model.add_cpds(EQ, Tor, Bur, AR)
print(model.check_model())
```

```python
infer = VE(model)
```

```python
What_is_the_probability_that_it_is_earthquake_given_the_alarm_is_ringing = infer.query(
    ['EarthQuake'],{'AlarmRinging':0})
print(What_is_the_probability_that_it_is_earthquake_given_the_alarm_is_ringing)
```

```python
What_is_the_probability_that_it_is_burglary_given_the_alarm_is_ringing = infer.query(
    ['Burglary'],{'AlarmRinging':0})
print(What_is_the_probability_that_it_is_burglary_given_the_alarm_is_ringing)
```

```python
What_is_the_probability_that_alarm_is_ringing_if_both_burglary_and_earthquake_happens = infer.query(
    ['EarthQuake','Burglary'],{'AlarmRinging':0})
print(What_is_the_probability_that_alarm_is_ringing_if_both_burglary_and_earthquake_happens)
```
