```python
class CyberSecurityGame:

    def __init__(self):
        self.security_score = 50   # starting score

        self.attacks = {
            "Phishing": -10,
            "Malware": -15,
            "DDoS": -20
        }

        self.defenses = {
            "Firewall": +10,
            "Patch Update": +15,
            "User Training": +8
        }


    # Utility function
    def utility(self, score):
        return score


    # Minimax with Alpha Beta
    def minimax(self, depth, is_attacker, score, alpha, beta):

        if depth == 0:
            return self.utility(score)

        if is_attacker:
            best = float('inf')

            for attack in self.attacks.values():

                new_score = score + attack
                value = self.minimax(depth-1, False, new_score, alpha, beta)

                best = min(best, value)
                beta = min(beta, best)

                if beta <= alpha:
                    break

            return best

        else:
            best = -float('inf')

            for defense in self.defenses.values():

                new_score = score + defense
                value = self.minimax(depth-1, True, new_score, alpha, beta)

                best = max(best, value)
                alpha = max(alpha, best)

                if beta <= alpha:
                    break

            return best


    # Find best defense
    def best_defense(self):

        best_score = -float('inf')
        best_move = ""

        for defense, value in self.defenses.items():

            score = self.security_score + value

            move_score = self.minimax(2, True, score, -float('inf'), float('inf'))

            if move_score > best_score:
                best_score = move_score
                best_move = defense

        return best_move, best_score
```

```python
game = CyberSecurityGame()

print("Current Security Score:", game.security_score)

attack = input("Enter attack (Phishing/Malware/DDoS): ")

game.security_score += game.attacks[attack]

print("Security Score after attack:", game.security_score)

best_move, score = game.best_defense()

print("AI Recommended Defense:", best_move)
print("Predicted Security Score:", score)
```
